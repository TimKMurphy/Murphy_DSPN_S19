# Is Classroom Foreign Language Learning Supported by Incidental Learning?

### <center><b>Background:</b><br></center>

Older children and adults have a particularly difficult time precieving speech sounds not found in their native language. These difficulties often make learning a second language (L2) difficult, especially if the L2 speech sounds differ substantially from those in the learner's native language. As such, there is great deal of interest from both researchers and L2 instructors on how to best address lanuguage learners' difficulties with the perception of L2 speech sounds. To properly address these issues, it is important to first have an understanding of how speech sounds are precieved by human beings. A perspective put forth by Dr. Lori Holt and colleagues suggests that such speech sound perception depends on general auditory processing mechanisms, specifically those processes which use certain acoustic features in the categorization of sounds. Evidence from their previous work suggests that difficulties in categorizing speech sounds the weight given to specific acoustic features used in the classification process. Thus, changing how an L2 learner utilizes acoustic features to categorize specific speech sounds could do much to improve their perception of L2 speech sounds. However, such acoustic features are typically found in the higher order statistical structure of the sound and occur very quickly. Changing how individuals precieve certain sound categories via explicit instruction would prove incredibly difficult, if not impossible. That being said, research has shown that improvements in sound categorization, speech or otherwise, can be accomplished via incidental learning. Incidental learning is a type of learning in which an individuals attention is purposely drawn away from to be learned stimuli. A video game specifically designed by Dr Holt and colleagues to promote incidental learning of sound categories has been particularly successful in showing how incidental learning can improve L2 speech sound categorization specifically. However, these studies have only studied participants who have played the incidental learnign game over the course of one week. Additionally, the interaction between simultaneous L2 classroom instruction and incidental auditory training has yet to be studied. 

The goal of the current study is twofold. First, we will assess the impact of long term incidental training on individuals' learning of novel speech sounds. Second, we will assess the impact of long term incidental training with explicit classroom instruction on individuals' learning of novel speech sounds. Four Mandarin tonal groups will serve as the novel speech sounds which participants learned. Three groups of participants will be recruited: students currently enrolled in an introductory Mandarin course, individuals not enrolled in an introductory Mandarin course, and native Mandarin speakers. The participants in the intoductory Mandarin group will be further subdivided such that half this group will be assigned to the training condition and the others a control condition. All Controls will be assigned the to the training group while no native Mandarin speakers will be assigned to the training condition. Participants in the training condition will play the incidental learning video game for 1 hour a week over the course of 8 weeks. After training, these participants, as well as the control participants and native Mandarin speakers, will perform an artifical language learning task over 3 consecutive days. The artifical language participants will learn uses the same tonal groups as Mandarin. Participants' learning will be assessed via the number of correct responses made on a four alternative choice (4AFC) categorization task involving words from this novel language. The dependent variable of interest in this study will be the improvement individuals make in the 4AFC categorization task over the course of 3 consecutive days.

<b>Data files (raw)</b>
All datafiles are human readable (not binary).

1. Participant tracking sheet<br>
Simple Microsoft Excel sheet (.xlsx) <br>
This sheet contains participant IDs, training group assignments, and Pitch Perception
2. Pre-Study Survey<br>
Exported from Qualtrics as .csv file <br>
This sheet contains Native Mandarin Speaker and Enrolled in introductory Mandarin
3. Mandarin speech sound discrimination task output<br>
Exported from E-Prime as .edat2 file.
Each subject has one such file.
This file contains accuracy for each trial on this task. Information from this file will be used to calculate Mandarin speech sound discrimination - percent correct
4. Userlog <br>
Exported from MySQL database as CSV file.
This file contains participants' play time for every training game session and the date of the first training session. Play time data will be used to calculate Total training time. Date of first training session will be used to calculate Time in training phase.
5. Artifical Langauge - 4 Alternative forced choice task output<br>
Exported from E-Prime as .edat2 file.
Each subject has one such file.
This file contains accuracy for each trial on this task. Information from this file will be used to calculate Artifical Language 4 Alternative Forced Choice task - Percent correct (4AFC_acc)
