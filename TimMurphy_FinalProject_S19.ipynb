#  Does Incidental Learning of Non-Speech Auditory Analogs by Second Language Learners Promote the Acquisition of Mandarin Tone Groups?

### <b>Background</b><br>

Older children and adults have a particularly difficult time precieving speech sounds not found in their native language. These difficulties often make learning a second language (L2) difficult, especially if the L2 speech sounds differ substantially from those in the learner's native language (Wang et. al. (2003), Chandrasekaran et. al. (2010), SO & Best (2010)). Why non-native speech sounds are so difficult for many adults to learn is due in no small part to the exposure and experience adults have with their native language. Years of experience with a native language allows the adult brain to quickly and effortlessly identify cues in native language speech to categorize speech sounds. However, these native speech sound categories often become so entrenched that they interfere with the learning of non-native speech categories. 

Many studies have tried to direct learners attention to specific features in non-native speech sounds in order to overcome native speech sound interference, though most results have been mixed (Bradlow et. al. (1997), Jamieson et. al. 1989, Pisoni et. al. (1994), Pisoni et. al. 1982). Interesting, research by Dr. Lori Holt and colleagues has shown that new auditory categories can be formed by directing learners attention <i>away</i> from to-be-learned auditory categories (Wade & Holt (2005), Lim & Holt (2011), Liu & Holt (submitted)). This type of learning is known as incidental learning as it occurs when participants are engaged in a task which does not primarily dependent on explicit attention to the stimuli being learned. While most of this work has focused on learning non-speech auditory categories, recent work by Liu & Holt (submitted) investigated the efficacy of incidental learning for the purposes of learning non-native language speech categories, specifically Mandarin tonal groups. In addition to training a group participants on Mandarin speech incidentally, a second group of participants was trained on stimuli consisting of isolated Mandarin tones. These stimuli were drawn from recordings of real Mandarin speech but were stripped of all phonetic information. Results from this study showed that both groups were able to generalize to and learn novel Mandarin words with simmilar speed and accuracy. Taken together, this work suggest that incidental learning and the use of non-speech stimuli varying along acoustic dimensions important for the categorization of speech sounds, may provide a "back door" through which non-native speech sounds can be learned without interference from native language.

### <b>Current Study</b>

The goal of the current study was to assess incidental auditory training in individuals  enrolled in a CMU introductory Mandarin course. Participants in this "classroom intervention" group (n=9) played a videogame-like task for 1 hour a week over the course of 8 weeks. This task was developed specifically to induce incidental learning of auditory categories. After 8 weeks of training, participants returned to the lab to perform an artifical language learning task over 3 consecutive days. Participants' learning was assessed via the number of correct responses made on a four alternative choice (4AFC) categorization task involving words from this novel language. Accuracy on this task over the course of 3 days served as the main dependent variable of interests

Three additional groups were recruited to test several hypotheses. A "train only" group (n=7) was formed consisting of individuals not enrolled in introductory Mandarin. However, participants in this group were assigned to play the incidental learning videogame over an 8 week period, just like the classroom intervnetion group, and returned to complete 3 consecutive days of artifical language taks. Comparisons between this group and the classroom intervention group will be performed to determine if classroom instruction and incidental auditory training interact, leading to more robust learning of speech sound categories. 

In addition to the train only group, a class only group (n=1), consisting of individuals enrolled in a CMU introductory Mandarin course who have not played the videogame for 8 weeks, was formed as well. Native Mandarin speakers (n=3) were also recruited to perform the artifical language tasks only. Due to the relatively small sizes of these groups, analyses of these participants performance will not take place in the analyses below.

## Hypotheses

### Hypothesis 1:<br>

The first hypothesis pertained to participants' baseline Mandarin perceptual abilites, specifically with regards to the Mandarin Speech Sound Discriminiation (MSSD) task. Because the stimuli heard on this task can be differentiated between via simple auditory cues, it was hypothesized that participants in the classroom intervention group and in the training only group would show no differences in MSSD accuracy. To test this hypothesis, mixed effects logistic regression was used. Introductory Mandarin classroom enrollment served as the main variable of interest, while pitch discrimination was included as a covariate. A random effect term was included to account for random variance of the intercept between indviduals.


Stated formally: <b>The probability of achieving higher accuracy on the Mandarin Speech Sound Discrimination task will not differ between the classroom intervention group and the training only group.</b> <br><br>

___


### Hypothesis 2:<br>
The second hypothesis pertained to participants' baseline Mandarin tone categorization abilites, specifically with regards to the Mandarin 4-item alternative forced choice (M4AFC) task. Unlike the MSSD task, it was hypothesized that participants enrolled in Introductory Mandarin class will perform better on the M4AFC. This was due simply to the fact that those participants in the class have knowledge of the existence of Mandarin tones, whereas those not enrolled in the class may not. As with the MSSD hypothesis, mixed effects logistic regression was used. Introductory Mandarin classroom enrollment served as the main variable of interest, while pitch discrimination was included as a covariate. A random effect term was included to account for random variance of the intercept between indviduals.

Stated formally: <b> The probability of achieving higher accuracy on the Mandarin 4-item alternative forced choice task will be greater in the classroom intervention group compared to the training only group.</b>

___


### Hypothesis 3:<br>
The thid hypothesis pertained to participants' performance on the artifical tonal language tasks, specifically accuraccy on 4-item Alternative Forced Choice (T4AFC) administered on all 3-days of post-training evaluation. It was hypothesized the classroom intervention group would outperform the training only group due to more exposure to the tonal groups in a classroom setting. Once again, mixed effects logistic regression was used. Introductory Mandarin classroom enrollment served as the main variable of interest, while pitch discrimination, training time on the incidental learnign video game, and day of testing were included as a covariates. A random effect term was included to account for random variance of the intercept between indviduals.

Stated formally: <b> The probability of achieving higher accuracy on the artifical tonal language 4-item aternative will be greater in the classroom intervention group compared to the training only group.</b>
    
___


### Quadratic Discriminate Analysis:<br>
In addition to accuracy measures on the T4AFC, information pertaining to the type of errors made by participants in the T4AFC tasks is available as well. There errors include "compliment" errors, which were responses made to an item in the same tone group as the target, "rhyme" errors, which were responses made to an item with the same phonetic information as the target, "distractor" errors, which were reponses made to an item sharing neither tonal nor phonetic information, and finally "misses" which were simply non-responses. A <b>Quadratic Discriminate Anlaysis</b> will be performed to see if pitch discrimination, training time on the incidental learning video game, and day of testing could be used to predict the type of responses made by participants. (NOTE: Class cannot be included in this analyses as it would violate the assumption of multivariate normality)

## Extracting Gameplay Time

Gameplay times are variable between participants in this study. How the amount of gameplay time effects TOFU learning is a very important question which is addressed in hypothesis 3.

While participants do self-report gameplay times, the server which hosts the Neuraltone game tracks the timing of events in the game relative to when the participant began gameplay. Getting the total gameplay time for a participant involves finding the last event entered for a particular game session. Each game session recieves a unique ID which allows for each game session to be easily parsed from the massive dataset containing all of the logged events for every Neuraltone game ever played. Determining which participant played which game is accomplished by looking at another table which connects game IDs to each participants user ID, which is assigned to each player by the server ({date}_ user_gameid.csv). Unfortunately, the user ID assigned by the server does not match with the participant study ID assigned by the study coordinators; However, a link between user ID and the study ID is kept in a secure file which uses email address to make the connection. User IDs are coded into the participant tracking sheet.

Using the user IDs from the tracking sheet, the script below finds each of the game IDs associated with that user ID. From there, corresponding game session data is extracted from the gameplay log file ({date}_ loguser_activity.csv). From this extracted data, the last recorded event time is taken. The last event time from each of these game sessions is then summed to yeild the total amount of gameplay for each participant over an 8-week period.

As of right now, the sheer size of the time logging table has prevent a proper data auditing technique to be implemented in this script. Right after the cell containing the gameplay time extraction scipt is another script which compares participant self-report gameplay time to gameplay times extracted from the server. While these two times tend to be close to one another, there are certainly some participants who have a sizeable gap between the server time and the reported tiimes. Whether the participant or script itself is at fault is not entirely clear at this time. However, it is believed that the self-report times are most likely the culprit, especially due to the fact that participant compensation increases with increased gameplay time.

```r
get_game_ids <- function(ID_ind, id_lookup)
    {
    part_gameids=game_userid$id[which(id_lookup$User_ID[ID_ind]==game_userid$user_id)]
    return(part_gameids)
    }

get_game_time_data <- function(game_id)
    {
        time_data=game_logs$time_elapsed[which(game_logs$game_id==game_id)]
        time_data=tail(time_data, n=1)
        if(length(time_data)==0) #if gameid does not exist, will get numeric(0). need to make 0
            {
            time_data=0
            }
        return(time_data)
    }

ID_part_user=c("ID", "User_ID")
User_id_lookup=na.omit(MASTER[ID_part_user])

num_users=nrow(User_id_lookup)
part_total_game_time=vector()

for(i in 1:num_users)
    {
    game_ids_4_part=get_game_ids(i, User_id_lookup)
    
    game_time_summer=0
    num_games=length(game_ids_4_part)
    for(j in 1:num_games)
        {
        game_time_data=get_game_time_data(game_ids_4_part[j])
        game_time_summer=game_time_summer+game_time_data
        }
    part_total_game_time[i]=game_time_summer
    }

#convert seconds to minutes, divide by 60
game_time_mins=part_total_game_time/60

User_id_lookup=cbind(User_id_lookup, game_time_mins)
MASTER_match_index=match(User_id_lookup$ID, MASTER$ID)

#create time vector to cbind with MASTER, then cbind
lookup_inc=1
train_time=vector()
for (h in 1:nrow(MASTER))
{
        if (MASTER$ID[h] %in% User_id_lookup$ID)
            {
            train_time[h]=game_time_mins[lookup_inc]
            lookup_inc=lookup_inc+1
            }
        else
            {
            train_time[h]=NA
            }
}

ID_list=data.frame(MASTER$ID)
colnames(ID_list)<-c("ID")
train_time=data.frame(train_time)
train_time=cbind(ID_list, train_time)
MASTER=full_join(MASTER, train_time, by="ID")
```

## Data Extraction Functions
The cell below contains three functions: MSSD_data_ripper, M4AFC_data_ripper, and T4AFC_data_ripper. All of these functions operate on E-prime output data. E-prime uses a relatively esoteric text file format. Luckily, there exists an r library, known as rprime, developed for extracting output data from E-prime text files.

#### MSSD_data_ripper()

This function extracts accuracy data from the Mandarin Speech Same Different task (MSSD). Because responses on this task consist only of "Same" or "Different", the "Response.ACC", or response accuracy field, is the only field which needs to be extracted. As data in this field are simply 1's and 0's, converting these values into qualitiative "Correct" and "Incorrect" responses was relatively easy.

#### M4AFC_data_ripper()

This function extracts accuracy data from the Mandarin 4-item Alternative Forced Choice task (M4AFC). Unlike the previous task set up, there are a total of 4 different response participants can give on the M4AFC task (one for each Mandarin tone group). For whatever reason, E-prime did not include an accuracy field as it had in the MSSD output. However, both participant responses and the correct responses were included as fields in the output file. Therefore, both the participant response and correct response were extracted. Accuracy was then determined by comparing the two matrices. An accuracy vector was built which indicated matches between the two fields with a 1, and non-matches with a 0. From here "Correct" and "Incorrect" responses were classified.

#### T4AFC_data_ripper()

This function extracts accuracy data from the TOFU 4-item Alternative Forced Choice task (T4AFC). It differs significantly from the previous two functions as the only participant measures collected on this task were continuous mouse position and position of mouse during a click. Of importance to this analysis is the clicking location only. Using a lookup table provided by Co-PI Seth Wiener (TOFU-cmu-AOIs-day{number}.csv), the function could determine which symbol the participant had clicked during a particular trial. Then, using a matching procedure similar to the one outline in the M4AFC function above, the function could determine the accuracy of a response for a given trial. Rather than "Correct" or "Incorrect", responses from this task were classified as "Correct", "Incorrect Consistent", "Incorrect Rhyme", "Incorrect Distraction", or "Incorrect Missing". For an explanation of these terms, see the study design above

It should be noted that, given the importance of screen position on this task, the exact screen resolution used during this task was extremely important. At the beginning of this function is code which checks an E-prime output file that contains information relating to screen resolution size for a given run. This was an XML and therefore required a specialized XML library to parse.

```r
#Mandarin speech same different task

MSSD_data_ripper<-function(pID, dir)
    {
    MSSD.file=paste(sep="", dir, pID, "/AXtone-", pID, "-1.txt" )
    MSSD=to_data_frame(FrameList(read_eprime(MSSD.file)))
    #grab accuracies
    MSSD_acc_TF=as.logical(as.numeric((na.omit(MSSD$Response.ACC))))
    MSSD_acc=rep("Incorrect", length(MSSD_acc_TF))
    MSSD_acc[MSSD_acc_TF]<-"Correct"
    return(MSSD_acc)
    }




#Mandarin 4AFC

M4AFC_data_ripper<-function(pID, dir)
    {
    M4AFC.file=paste(sep="", dir, pID, "/4AFC-", pID, "-1.txt" )
    M4AFC=to_data_frame(FrameList(read_eprime(M4AFC.file)))
    #grab response, replace NA with 0
    response=as.numeric(na.omit(M4AFC$Response.RESP))
    response[is.na(response)]<-0
    #grab correct response, omit NA (which shouldnt be in there to begin with)
    corr_response=as.numeric(na.omit(M4AFC$Response.CRESP))
    #get accuracy by finding matches between resp and correct resp
    M4AFC_acc<-corr_response==response
    #build matrix with zeros then apply accuracy matrix to insert 1's
    to_return=rep("Incorrect",length(M4AFC_acc))
    to_return[which(M4AFC_acc==1)]<-"Correct"
    
    #divide by length of corr_response (again, should have no NA) to get percent accuracy
    return(to_return)
    }

#TOFU 4AFC acc ripper

T4AFC_data_ripper<-function (pID, directory, day)
{
    #Check which day this is
    if (day==1)
    {
        tofu_4afc_aoi=tofu_4afc_aoi_d1
    }
    else if (day==2)
    {
        tofu_4afc_aoi=tofu_4afc_aoi_d2
    }
    else if (day==3)
    {
        tofu_4afc_aoi=tofu_4afc_aoi_d3
    }
    
    #EXTREMELY IMPORTANT!
    #This code gets the screen resolution size for the task
    #this info is needed because multiple screens were used
    exp_report=paste(sep="",directory, pID ,"/","Day1-",pID,"-ExperimentAdvisorReport.xml")
    doc=xmlParse(exp_report)
    nodes_of_interest=xpathApply(doc, "//Device/Property/PropertyValue",xmlValue)
    screen_x=as.numeric(nodes_of_interest[4])
    screen_y=as.numeric(nodes_of_interest[5])
    
    mid_x=screen_x/2 #horizontal mid
    mid_y=screen_y/2 #vertical mid
    
    
    t4AFC=paste(sep="",directory, pID,"/Day1-", pID, ".txt" )
    t4AFC_file=to_data_frame(FrameList(read_eprime(t4AFC)))
    X=as.numeric(na.omit(t4AFC_file$XTrackedFinal))
    Y=as.numeric(na.omit(t4AFC_file$YTrackedFinal))
    ans_total=length(X)
    ans_holder=vector()
    for (j in 1:ans_total)
        {
            if (X[j]<mid_x) #left side
                {
                if (Y[j]<mid_y)
                    {
                    ans_holder[j]=1 #left top
                    }
                if (Y[j]>mid_y)
                    {
                    ans_holder[j]=4 #left bottom
                    }
                }
            if (X[j]>mid_x) #right side
                {
                if (Y[j]<mid_y)
                    {
                    ans_holder[j]=2 #right top
                    }
                if (Y[j]>mid_y)
                    {
                    ans_holder[j]=3 #right bottom
                    }
                }        
        }
    t4afc_acc_target<-ans_holder==na.omit(tofu_4afc_aoi$targetAOI)
    t4afc_acc_comp<-ans_holder==na.omit(tofu_4afc_aoi$competitorAOI)
    t4afc_acc_rhyme<-ans_holder==na.omit(tofu_4afc_aoi$rhymeAOI)
    t4afc_acc_dist<-ans_holder==na.omit(tofu_4afc_aoi$distAOI)
    t4afc_response=rep("Incorr_Miss",ans_total)
    t4afc_response[t4afc_acc_target]="Correct"
    t4afc_response[t4afc_acc_comp]="Incorr_Competitor"
    t4afc_response[t4afc_acc_rhyme]="Incorr_Rhyme"
    t4afc_response[t4afc_acc_dist]="Incorr_Distract"
    
    #t4afc_acc=sum(t4afc_acc) NO SUMMING
    t4afc_bundle=c(t4afc_response)
    return(t4afc_bundle)
}
```

## Extracting Mandarin Tasks Data
The script below runs increments through each row (participant) of the MASTER table and extracts responses on the Mandarin Speech Same Different task and the Mandarin 4-item alternative forced choice (M4AFC) task.

The if-then statements in this code check to see if the participant was either dropped from the study, or is incomplete. These participants will recieve an NA for the MSSD and M4AFC. The later designation is important because some of the individuals listed in the data sheet are still enrolled in the study. Once they have finished, their remaining data files are put into the file directory and the one in the "incomplete" column is dropped for that participant. This step was added to simplify the addition of participants who have yet to complete.

For data from participants who have completed the study, the script simply calls the MSSD_data_ripper and M4AFC_data_ripper functions. These functions return two variables. The first is a vector containing the correct/incorrect responses from each task. The second is the total number of responses from each task. This measure will be used specifically for data auditing purposes. For the MSSD task, a total of 196 responses should be returned, whereas for the M4AFC task, 200 responses will be returned.

```r
num_subjs=length(MASTER$ID)

drops=MASTER[which(MASTER$drop==1),]
drops=drops$ID

incompletes=MASTER[which(MASTER$incomplete==1),]
incompletes=incompletes$ID

MSSD_total=vector()
MSSD_acc=matrix(, nrow = num_subjs, ncol = 196) #row = particiapnt, col = response on trial
        
M4AFC_total=vector()
M4AFC_acc=matrix(, nrow = num_subjs, ncol = 200)  #row = particiapnt, col = response on trial
        

for (k in 1:num_subjs)
    {
    pID=toString(MASTER$ID[k])   
    #print(pID)
    #for drops, set all values to NA
    if (pID %in% drops)
        {
        MSSD_total[k]=NA
        MSSD_acc[k]=NA
        
        M4AFC_total[k]=NA
        M4AFC_acc[k]=NA
        }
    #GRAB MSSD and 4AFC
    else
        {
        #npID = numeric representation of pID, which is a string
        npID=as.numeric(pID)
        #Mandarin speakers have MSSD & M4AFC data, however they begin 41.. not 40, need to check
        if (npID>4000)
            {
            pID=toString(npID+100)
            }
        
        MSSD_acc[k,]=MSSD_data_ripper(pID, directory)
        MSSD_total[k]=length(MSSD_acc[k,])
        
        #Audit - less than 196 response? less than chance (50% - 98) performance?
        if (MSSD_total[k]!=196)
            {
            MSSD_total_warning=paste("WARNING: MSSD total does not equal 196 for participant ", pID," ", toString(MSSD_total[k]))
            print(MSSD_total_warning)
            }
        chance_check=length(which(MSSD_acc[k,]=="Correct"))
        if (chance_check<98)
            {
            MSSD_chance_warning=paste("WARNING: MSSD accuracy less than chance ", pID," - ", toString(chance_check))
            print(MSSD_chance_warning)
            }
            
        M4AFC_acc[k,]=M4AFC_data_ripper(pID, directory)
        M4AFC_total[k]=length(M4AFC_acc[k,])
        
         #Audit - less than 200 response? less than chance (25% - 50) performance?
        if (M4AFC_total[k]!=200)
            {
            M4AFC_total_warning=paste("WARNING: M4AFC total does not equal 200 for participant ", pID," - ", toString(M4AFC_total[k]))
            print(M4AFC_total_warning)
            }
        chance_check=length(which(M4AFC_acc[k,]=="Correct"))
        if (chance_check<50)
            {
            M4AFC_chance_warning=paste("WARNING: M4AFC acc less than chance", pID," ", toString(chance_check))
            print(M4AFC_chance_warning)
            }
        }
    }

ID_list=data.frame(MASTER$ID)
colnames(ID_list)<-c("ID")

#name MSSD variables and prep for tidyverse join
MSSD_acc=data.frame(MSSD_acc)
MSSD_resp_num=as.matrix(rep(1:196,1))
MSSD_resp_num=apply(MSSD_resp_num,1,FUN=toString)
MSSD_resp_name=as.matrix("MSSD")
MSSD_resp_name=apply(MSSD_resp_name,1, FUN=paste, MSSD_resp_num,sep="_")
colnames(MSSD_acc)<-MSSD_resp_name
MSSD_acc=cbind(ID_list,MSSD_acc)

#name M4AFC variables and prep for tidyverse join
M4AFC_acc=data.frame(M4AFC_acc)
M4AFC_resp_num=as.matrix(rep(1:200,1))
M4AFC_resp_num=apply(M4AFC_resp_num,1,FUN=toString)
M4AFC_resp_name=as.matrix("M4AFC")
M4AFC_resp_name=apply(M4AFC_resp_name,1, FUN=paste, M4AFC_resp_num,sep="_")
colnames(M4AFC_acc)<-M4AFC_resp_name
M4AFC_acc=cbind(ID_list,M4AFC_acc)

#join MASTER, MSSD on ID then MASTER, 4AFC
MASTER=full_join(MASTER, MSSD_acc, by="ID")
MASTER=full_join(MASTER, M4AFC_acc, by="ID")
#MASTER
```

## Extracting responses from the TOFU 4-item Alternative Forced Choice task

The if-then statements in this code check to see if the participant was either dropped from the study, or is incomplete. The later designation is important because some of the individuals listed in the data sheet are still enrolled in the study. Once they have finished, their remaining data files are put into the file directory and the one in the "incomplete" column is dropped for that participant. This step was added to simplify the addition of participants who have yet to complete.

Participants who have completed the study have T4AFC responses extracted for each day via the "T4AFC_data_ripper()" function (details on function mechanics described above). During the extraction process, a check is added to make sure responses for each day total 48 items. A warning is thrown if not. Data frames (variables d1,d2,d3) are created to temporarily store each collection of responses, the corresponding DAY varibale, and all of the participant's variables located in MASTER. After all three days of T4AFC responses are collected, they are binded by row to create a temporary data frame for all 3 days worth of responses for that participant.

Finally, all participants' responses (including drops and incompletes) are row bound to an empty data frame (variable all_resp). The resulting data table contains each participants' response to each trial on the T4AFC tasks across 3 days as seperate rows. In this sense, each response on the T4AFC task serve as an observation.


```r
ID_list=MASTER$ID
ID_list=as.matrix(ID_list)
colnames(ID_list)<-c("ID")
num_subj=length(ID_list)
all_resp=data.frame()
d1=matrix(,nrow=num_subj, ncol=48)
d2=matrix(,nrow=num_subj, ncol=48)
d3=matrix(,nrow=num_subj, ncol=48)

#this function takes the responses in matrix form after the script crawls through the data
#it is naming the columns for later merging purposes and adding corresponding IDs

t4afc_day_ta_prep <- function(day_ta, day) #day_ta, isn't that funny? no? :(
    {
            T4AFC_resp_name=as.matrix(rep("T4AFC_",48))
            T4AFC_resp_day=toString(day)
            T4AFC_resp_name=apply(T4AFC_resp_name,2,FUN=paste,sep="",T4AFC_resp_day)
            colnames(day_ta)<-T4AFC_resp_name
            day_ta=cbind(ID_list, day_ta)
            day_ta=data.frame(day_ta)
            return(day_ta)

    }


for (u in 1:num_subjs)
    {
    pID=levels(MASTER$ID)[u] 
        if (pID %in% drops | pID %in% incompletes)
            {
            t4afc_response=rep(NA,48)
            d1[u,]=t4afc_response
            t4afc_response2=rep(NA,48)
            d2[u,]=t4afc_response2
            t4afc_response=rep(NA,48)                                     
            d3[u,]=t4afc_response3
            }
        else
            {
            # addition of 100, 200, 300 to participant ID needed to access T4AFC files on different days. For examples,
            # if 1001 has all data from all three days, their number on day 1 becomes 1101, day 2 1201, and day 3 1301
            num_ID=as.numeric(levels(MASTER$ID)[u])
            pID1=toString(num_ID+100)
            pID2=toString(num_ID+200)
            pID3=toString(num_ID+300)
            
            
            t4afc_response1=T4AFC_data_ripper(pID1,directory,1)
            if (length(t4afc_response1)!=48)
                {
                missing_data_warning=paste(pID1, " missing data!", " count=", length(t4afc_response1))
                print(missing_data_warning)
                }
            d1[u,]=t4afc_response1
            
            t4afc_response2=T4AFC_data_ripper(pID2,directory,2)
            if (length(t4afc_response2)!=48)
                {
                missing_data_warning=paste(pID1, " missing data!", " count=", length(t4afc_response2))
                print(missing_data_warning)
                }
            d2[u,]=t4afc_response2
            
           t4afc_response3=T4AFC_data_ripper(pID3,directory,3)
            
           if (length(t4afc_response3)!=48)
                {
                missing_data_warning=paste(pID1, " missing data!", " count=", length(t4afc_response3))
                print(missing_data_warning)
                }
            d3[u,]=t4afc_response3
            }
    }


d1=t4afc_day_ta_prep(d1,1)
d2=t4afc_day_ta_prep(d2,2)
d3=t4afc_day_ta_prep(d3,3)



MASTER=full_join(MASTER, d1, by="ID")
MASTER=full_join(MASTER, d2, by="ID")
MASTER=full_join(MASTER, d3, by="ID")
```

# Hypothesis 1:

<center><b>Mandarin Speech Sound Discrimination</b></center><br> 
The first hypothesis pertained to participants' baseline Mandarin perceptual abilites, specifically with regards to the Mandarin Speech Sound Discriminiation (MSSD) task. Because the stimuli heard on this task can be differentiated between via simple auditory cues, it was hypothesized that participants in the classroom intervention group and in the training only group would show no differences in MSSD accuracy. To test this hypothesis, mixed effects logistic regression was used. Introductory Mandarin classroom enrollment served as the main variable of interest, while pitch discrimination was included as a covariate. A random effect term was included to account for random variance of the intercept between indviduals.


Stated formally: <b>The probability of achieving higher accuracy on the Mandarin Speech Sound Discrimination task will not differ between the classroom intervention group and the training only group.</b> <br><br>

```r
library(plyr) #helpful with factor renaming
#build seperate tables for MSSD and M4AFC only data
MSSD_data=select(MASTER, starts_with("ID"),starts_with("Mand_speak"),starts_with("drop"),starts_with("incomplete"), starts_with("Pitch"), starts_with("Class"),starts_with("MSSD"))

#remove drops, incompletes, and Mandarin native speakers
MSSD_data=MSSD_data %>% filter(is.na(drop))
MSSD_data=MSSD_data %>% filter(is.na(incomplete))
MSSD_data=MSSD_data %>% filter(Mand_speak==0)

MSSD_only=select(MSSD_data,starts_with("MSSD"))
#because MSSD bins are stored as factors, need to convert to char before casting
MSSD_only <- data.frame(lapply(MSSD_only, as.character), stringsAsFactors=FALSE)
MSSD_only=gather(MSSD_only,key="MSSD_bin", value="MSSD_acc")

independent=select(MSSD_data, c(1:10))
MSSD_analysis=cbind(MSSD_only,independent)
MSSD_analysis$Class=as.factor(MSSD_analysis$Class)
MSSD_analysis$Class=mapvalues(MSSD_analysis$Class, from=c("0","1"), to=c("No Class","Class"))
MSSD_analysis$MSSD_only=as.factor(MSSD_analysis$MSSD_acc)
MSSD_analysis$MSSD_only=relevel(MSSD_analysis$MSSD_only, "Incorrect") #relevel so that Incorrect is 0, correct is 1

MSSD_plot=select(MSSD_analysis, starts_with("MSSD_only"), starts_with("Class"))
Class_freq=count(MSSD_plot$MSSD_only[MSSD_plot$Class=="Class"])
No_class_freq=count(MSSD_plot$MSSD_only[MSSD_plot$Class=="No Class"])
Class_freq$freq=Class_freq$freq/length(MSSD_plot$MSSD_only[MSSD_plot$Class=="Class"])
Class_freq=cbind(Class_freq, c("Class","Class"))
colnames(Class_freq)<-c("response", "perc","Class")
No_class_freq$freq=No_class_freq$freq/length(MSSD_plot$MSSD_only[MSSD_plot$Class=="No Class"])
No_class_freq=cbind(No_class_freq, c("No Class","No Class"))
colnames(No_class_freq)<-c("response", "perc","Class")

MSSD_plot=rbind(Class_freq,No_class_freq)
ggplot(MSSD_plot, aes(x=Class, y=perc, fill=response))+geom_bar(stat='identity', position='dodge')+ylab("Percentage")+
geom_text(aes( label = scales::percent(MSSD_plot$perc),y= MSSD_plot$perc ), position = position_dodge(0.9), vjust=-0.2) 
#sort by ID for point graph below
MSSD_analysis <- MSSD_analysis[order(MSSD_analysis$ID),]




#
#analysis
#



contrasts(MSSD_analysis$MSSD_only)
MSSD.melr_model=glmer(MSSD_only~Class+Pitch_Discrimination+(1|ID), data=MSSD_analysis, family="binomial")
MSSD.lr_model=glm(MSSD_only~Class+Pitch_Discrimination, data=MSSD_analysis, family="binomial")
anova(MSSD.melr_model, MSSD.lr_model)

summary(MSSD.melr_model)
MSSD.predict = data.frame(predict(MSSD.melr_model, type = "response"))
colnames(MSSD.predict)<- c("predicted_response")
n_MSSD = nrow(MSSD_analysis)
MSSD.predict$index = seq(1, n_MSSD)
confint(MSSD.melr_model)

ggplot(MSSD.predict, aes(index, predicted_response)) + geom_point(aes(color=MSSD_analysis$Class)) + xlab('observation') + ylab('predicted response') 

threshold = 0.8

MSSD.predict$predicted_binary=rep("Incorrect",n_MSSD)
MSSD.predict$predicted_binary[MSSD.predict$predicted_response>threshold]="Correct"

MSSD.confuse = data.frame(MSSD.predict$predicted_binary, MSSD_analysis$MSSD_only)
colnames(MSSD.confuse) = c('predicted', 'actual')

table(MSSD.confuse)
mean(MSSD.confuse$predicted == MSSD.confuse$actual)
```

# Hypothesis 2 - Results

<b>Mandarin 4-item Alternative Forced Choice</b><br>

After fitting a mixed effects logistic regression model, it was found that the log probability of responding correctly on the Mandarin 4-item Alternative Forced Choice task increased if participants were enrolled in an Introductory Mandarin compared to those who were not:<br>

$\hat\beta_{Class}$ =  <b>1.98 (SD=.44),</b>       95% CI <b>[1.06,2.92]</b>,      p=<b>7.8e-06</b>

Given this extremely small p-value, it is very likely that enrollment in a Mandarin class can lead to greater accuracy in categorizing the tones of Mandarin words.

Additonally, log probability of responding correctly on the Mandarin 4-item Alternative Forced Choice was also found to covary with Participant's Pitch Discrimination abilities:<br>

$\hat\beta_{Class}$ =  <b>-0.054 (SD=.019),</b>       95% CI <b>[	-0.093,	-0.015]</b>,      p=<b>0.0036</b>



The mixed effect logistic regression model built from this analyses was found to have a <b>prediction accuracy of about 70%.</b>




# Hypothesis 3:

<b>Artifical Tonal Language 4-item Alternative Forced Choice</b><br>
The thid hypothesis pertained to participants' performance on the artifical tonal language tasks, specifically accuraccy on 4-item Alternative Forced Choice (T4AFC) administered on all 3-days of post-training evaluation. It was hypothesized the classroom intervention group would outperform the training only group due to more exposure to the tonal groups in a classroom setting. Once again, mixed effects logistic regression was used. Introductory Mandarin classroom enrollment served as the main variable of interest, while pitch discrimination, training time on the incidental learnign video game, and day of testing were included as a covariates. A random effect term was included to account for random variance of the intercept between indviduals.

Stated formally: <b> The probability of achieving higher accuracy on the artifical tonal language 4-item aternative will be greater in the classroom intervention group compared to the training only group.</b>
    



```r
T4AFC_data=select(MASTER, starts_with("ID"),starts_with("drop"),starts_with("incomplete"), starts_with("Pitch"),starts_with("Mand_speak"), starts_with("Class"),starts_with("train_time"),starts_with("T4AFC"))

T4AFC_data=T4AFC_data %>% filter(is.na(drop))
T4AFC_data=T4AFC_data %>% filter(is.na(incomplete))
T4AFC_data=T4AFC_data %>% filter(Mand_speak==0)

T4AFC_only=select(T4AFC_data,starts_with("T4AFC"))
#because T4AFC bins are stored as factors, need to convert to char before casting
T4AFC_only <- data.frame(lapply(T4AFC_only, as.character), stringsAsFactors=FALSE)
T4AFC_only=gather(T4AFC_only,key="T4AFC_bin", value="T4AFC_acc")


n_observations=nrow(T4AFC_data)*3*48
T4AFC_only=T4AFC_only[1:n_observations,2] 
independent=select(T4AFC_data, c(1:10))
T4AFC_analysis=cbind(independent, T4AFC_only)


# add in days of testing
# create vectors containing day of testing number (make sure numeric), repeat day # by 48 (num observation per T4AFC)
day1=rep(0,48)
day2=rep(1,48)
day3=rep(2,48)
day_of_test=append(day1,day2)
day_of_test=append(day_of_test,day3)
day_of_test=rep(day_of_test, nrow(T4AFC_data))
T4AFC_analysis=cbind(T4AFC_analysis, day_of_test)
T4AFC_analysis$Class=as.factor(T4AFC_analysis$Class)
T4AFC_analysis$Class=mapvalues(T4AFC_analysis$Class, from=c("0","1"), to=c("No Class","Class"))

#
#PLOT DATA
#
 

T4AFC_responses=rep("Incorrect",nrow(T4AFC_analysis))
T4AFC_corrects=grep("Correct",T4AFC_analysis$T4AFC_only)
T4AFC_responses[T4AFC_corrects]<-"Correct"
T4AFC_responses=factor(T4AFC_responses)
T4AFC_4_logreg=select(T4AFC_analysis, starts_with("ID"), starts_with("Pitch_Discrimination"),starts_with("train_time"),starts_with("Class"),starts_with("day_of_testing"))
T4AFC_4_logreg=cbind(T4AFC_4_logreg, T4AFC_responses)
T4AFC_4_logreg$T4AFC_responses=relevel(T4AFC_4_logreg$T4AFC_responses, "Incorrect") #relevel so that Incorrect is 0, correct is 1
contrasts(T4AFC_4_logreg$T4AFC_responses)

#convert mins to hour -> done for mixed model, which could not converge with large values for minutes
train_time_hours=(T4AFC_analysis$train_time)/60
T4AFC_4_logreg=cbind(T4AFC_4_logreg, train_time_hours)
T4AFC_4_logreg<-T4AFC_4_logreg[order(T4AFC_4_logreg$ID),]


#analysis
T4AFC.melr_model=glmer(T4AFC_responses~Class+Pitch_Discrimination+train_time_hours+day_of_test+(1|ID), data=T4AFC_4_logreg, family="binomial")
T4AFC.lr_model=glm(T4AFC_responses~Class+Pitch_Discrimination+train_time_hours+day_of_test, data=T4AFC_4_logreg, family="binomial")
anova(T4AFC.melr_model, T4AFC.lr_model)

summary(T4AFC.melr_model)
confint(T4AFC.melr_model)



T4AFC_prob_df = data.frame(predict(T4AFC.melr_model, type = "response"))
colnames(T4AFC_prob_df) = c('predicted_prob')
n_observations = nrow(T4AFC_prob_df)
T4AFC_prob_df$index = seq(1, n_observations) 
T4AFC_prob_df$ID = T4AFC_4_logreg$ID[1:n_observations]
ggplot(T4AFC_prob_df, aes(index, predicted_prob)) + geom_point(aes(color=T4AFC_prob_df$ID))+ xlab('observation') + ylab('predicted response probability')

threshold = 0.5 #binarizing threshold 

# First make a list of "Downs"
T4AFC_prob_df$predicted_binary=rep("Incorrect",n_observations)

# Then use the probability output to label the up days. Let's use a threshold of 50% probability. 
T4AFC_prob_df$predicted_binary[T4AFC_prob_df$predicted_prob>threshold]="Correct" #find the rows that have prob. > threshold and cast as 'up'
# Now let's look at the prediction accuracy
confusion_df = data.frame(T4AFC_prob_df$predicted_binary, T4AFC_4_logreg[1:n_observations,5])
colnames(confusion_df) = c('predicted', 'actual')
table(confusion_df)

mean(confusion_df$predicted == confusion_df$actual)
```

# Hypothesis 3 - Results

<b>Artifical Language 4-item Alternative Forced Choice</b><br>

After fitting a mixed effects logistic regression model, it was found that the log probability of responding correctly in the Artifical Language 4-item Alternative Forced Choice tasks decreased if participants were enrolled in an Introductory Mandarin compared to those who were not:<br>

$\hat\beta_{Class}$ =  <b> -.587 (SD=.21),</b>       95% CI <b>[-.19, .60]</b>,      p=<b>.27</b>

However, given the relatively large p-value, it would be inappropriate to conclude that enrollment in a Mandarin class lead to better tone categorization on this task. <b>Thus, we do NOT reject the null hypothesis</b>.

Day of test is the only covariate which appears to have any increase on the probability participants will respond correctly.<br><br>
$\hat\beta_{Day of Test}$ =  <b>.22 (SD=.054),</b>       95% CI <b>[.12, .33]</b>,      p=<b>3.21e-5</b><br><br>
In the chart above, you can see the increase in this probability by looking at the predicted responses of participants (points which are uniquely color) as they move to the right. Notice that they tend to move up in steps. This correspondense to the different days participants were tested on the artifical language <br><br>

Other than day of test, there does not appear to be any other factors which were strongly associated with the log probability of responses on this task either. That being said, the direction of these estimates are all positive, which is what was expected.
<br><br>
$\hat\beta_{Pitch Discrimination}$ =  <b>.0030 (SD=.0076),</b>       95% CI <b>[-.013,.019]</b>,      p=<b>0.70</b><br>
$\hat\beta_{Train Time Hours}$ =  <b>0.015 (SD=.022),</b>       95% CI <b>[-.031, .062]</b>,      p=<b>0.49</b><br>

# Discussion

In this study, two groups of participants were compared. The classroom intervention group consisted of individuals currently enrolled in a CMU introductory Mandarin class who also performed 8 weeks of incidental auditory training on a videogame task. They were compared to a training only group consisting of individuals who were not enrolled in an introductory Mandarin class, though they did perform 8 weeks of incidental auditory training on the same videogame task.

Analyses of the two Mandarin tasks at the initial lab visit strongly indicated an advantage of Mandarin classroom instruction for both Mandarin speech sound discrimination and categorization of Mandarin tone. The latter was not surprising as individuals who were not enrolled in the training group most likely did not have knowlege of these tonal groups like the participants who were in the classroom intervention group. However, it was slightly surprising to see that participants in the classroom interevention had a higher probability of responding correctly on the speed sound discrimination task. The original logic behind our hypothesis was that participants would merely need to attend to acoustic features of the sound to differentiate between Mandarin words. However, it is possible that by rely soley on acoustic features, participants in the training only group may have made more "Different" responses than the classroom interevention group, leading to more incorrect responses. The classroom intervention group may have also already had exposure to some of the basic words which were presented during the task. Thus, differences between the utterances of Mandarin speakers might have been ignored by these participants.

Most surprising was the lack of any difference between the classroom intervention group and the training only group with regards to the probability of responding correctly on the artifical language 4-item alternative forced choice tasks. The fact the day of testing variable did seem to have a strong effect on correct response probability makes it unlikely that the stimuli was too hard for the participants to learn. The lack of an effect of gameplay on correct response probability was also rather surprising. That being said, it may be worthwhile to look at the regularity of gameplay time. It is possible that some individuals may have fallen off in their hours by the end of the study, causing learning decay. The server backend for the game does allow for a more nuanced measure of gameplay time, so this is an avenue definetly worth exploring.

Finally, it was rather disappointing that the quadratic linear discriminate analysis did not work. Why this occurred seems related to data I was hoping to test the model on. However, it is interesting to note that the largest error made by participants appears to be for stimuli which share the same tone category as the target. This is definetly an avenue I hope to explore as more data becomes available, especially from the native Mandarin speaker group and the introductory Mandarin group that will not train on the video game.
