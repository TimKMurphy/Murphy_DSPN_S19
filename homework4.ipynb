1. Consider the problem of classifying a binary response variable (i.e., $y \in \{1,0\}$). If there is no overlap in the values of X when y = 1 and when y=0,  such that there is a large “gap” between the two distributions of X values, then this is problematic for one of the classifiers discussed in class and the text. What classifier does this situation pose a problem for? Explain conceptually why this is a problem and compare it with another classifier approach that does not suffer this limitation.

**Answer:**
The "large gap" issue described above is a problem for logistic regression. This is a probelm for logistic regression because it leads to "unstable" estimates of model parameters. What this means is that, when classes are well seperated, multiple solutions exist for model parameters. Linear Discriminate Analysis, another kind of classifer, does not have problems with large sepertations between groups because it uses a discriminate function to define an optimal boundary between groups.

2. Compare logistic regression, LDA, and kNN classification approaches. Which are parametric which are non-parametric? For parametric models what functions do they assume? For non-parametric methods, how do the classifiers separate groups? How is the flexibility/bias tradeoff adjusted for each method?




**Answer:** 
Of the three classifier approaches, both logistic regression and LDA are parametric, whereas kNN is non-parametric. With regards to the parametric models, the logistic regression assumes a logistic function whereas LDA assumes a discriminate function. As is the case with other parametric models, there is a flexibility/bias tradeoff in this models. As the number of predictors $p$ increases relative to the number of $n$ observations, the model becomes very flexible, to the point where it could be overfitting the data. Conversly, the model could be biased if the number of predictors is low relative to the number of observations 

For the non-parametic model, kNN, a datapoint is classified by calculating the distance between this datapoint and all other points in the dataset. Once all distances have been calculated, the nearest $k$ datapoints are chosen. The datapoint is then classified as belonging to the class which contains a majority of its neighbors in the $k$ sized set. The value chosen for $k$ determines the flexibility (bias/variance tradeoff). Small $k$'s give tight fits for each group, to the point where they could overfit the data (too much variance). Large $k$'s do not give very tight fits, which does protect against overfitting, but could also bias which class an observation is assigned to.

3. What is the curse of dimensionality? Why is it especially problematic for kNN classification (i.e., why does kNN fail in high dimensional contexts)?


**Answer:**
The curse of dimensionality arises when the number of predictors (dimensions) $p$ is relatively large compared to the number of observations $n$ (can also be thought of as a high value for the ratio $p/n$). This is problematic for kNN because if $p/n$ is large, then an observation will likely not have enough nearby neighbors for classification. Thus, the model will need to go to greater distances to find more neighbors, which in turn increases the amount of noise (MSE) in the model.

4. Suppose that we take a data set, divide it into equally-sized training and test sets, and then try out two different classification procedures. First we use logistic regression and get an error rate of 20% on the training data and 30% on the test data. Next we use 1-nearest neighbors (i.e. K = 1) and get an average error rate (averaged over both test and training data sets) of 18%. Based on these results, which method should we prefer to use for classification of new observations? Why?

**Answer:** 
While the kNN method has a smaller average error rate between training and test data, it is important to note that, when $k=1$, kNN is essentially building a regression model of the data. Thus, error here is related to how well the function f(x) describes the relationship between some continuous variable x and a continuous variable y. Because y is continuous here, it is not suited for classification.

While the error rate on the logistic regression model increases from the training data error rate of 20% to a testing data error rate of 30%, it is at least classifying the continuous variable x into discrete variable y. Error now is not related to issues regarding the fit between continous variables x and y, but to errors fitting a continuous variable x to discrete variable y. For classification purposes, this is preferred.

```r
#(b) Use the full data set to perform a logistic regression with Direction as the response and the five lag variables plus Volume as predictors. 
#Use the summary function to print the results. 
#Do any of the predictors appear to be statistically significant? If so, which ones?

week.lr=glm(Direction ~ Lag1 + Lag2 + Lag3 + Lag4 + Lag5, data=Weekly, family=binomial)
summary(week.lr)
print("Based on a significance level of p<.05, only the coefficient for the intercept and Lag2 are statistically significant")

```

```r
#(c) Compute the confusion matrix and overall fraction of correct predictions. 
#Explain what the confusion matrix is telling you about the types of mistakes made by logistic regression.

threshold = .5
week.lr.predict=data.frame(predict(week.lr, type="response"))
colnames(week.lr.predict)=c("predict_prob")
size_week.lr=nrow(week.lr.predict)
week.lr.predict$predicted_binary=rep("Down", size_week.lr)
week.lr.predict$predicted_binary[week.lr.predict$predict_prob>threshold]="Up"
confuse_table=data.frame(week.lr.predict$predicted_binary, Weekly$Direction)
colnames(confuse_table)=c("predicted","actual")
table(confuse_table)

print("The confusion error tells us about the type of error so that we know whether the threshold we set is too high or too low. Here, it appears our threshold may be too low, as many of the Down directions are classified as up")
```

```r
#(d) Now fit the logistic regression model using a training data period from 1990 to 2008, with Lag2 as the only predictor. 
#Compute the confusion matrix and the overall fraction of correct predictions 
#for the held out data (that is, the data from 2009 and 2010).
threshold=.5

train=subset(Weekly, Year>1989|Year<2009)
test=subset(Weekly, Year<1990|Year>2008)
train.lr=glm(Direction ~ Lag2, data=o, family="binomial")
test_ind=nrow(test)
train.lr$predict_prob=predict(train.lr, test)
train.lr$predict_dir=rep("Down", test_ind)
train.lr$predict_dir[train.lr$predict_prob>threshold]="Up"
confuse_table_2=data.frame(train.lr$predict_dir, test$Direction)
colnames(confuse_table_2)=c("predicted","actual")
table(confuse_table_2)
mean(test$Direction==train.lr$predict_dir)
```

```r
#(e) Repeat (d) using LDA.
train=subset(Weekly, Year>1989|Year<2009)
test=subset(Weekly, Year<1990|Year>2008)
train.lda=lda(Direction ~ Lag2, data=o, family="binomial")
test_ind=nrow(test)
train.lda$predict_prob=predict(train.lda, test)

print("LDA")
table(train.lda$predict_prob$class, test$Direction)
mean(test$Direction==train.lda$predict_prob$class)

#(f) Repeat (d) using QDA.

train.qda=qda(Direction ~ Lag2, data=o, family="binomial")
test_ind=nrow(test)
train.qda$predict_prob=predict(train.qda, test)

print("QDA")
table(train.qda$predict_prob$class, test$Direction)
mean(test$Direction==train.qda$predict_prob$class)

#(g) Repeat (d) using KNN with K = 1.
library(class)
knn.pred=knn(data.frame(train$Lag2), data.frame(test$Lag2), train$Direction ,k=1)
print("KNN")
table(knn.pred, test$Direction)
mean(knn.pred==test$Direction)

#(h) Which of these methods appears to provide the best results on
#this data?

print("KNN provides the best results for this data")
```

```r
#(i) Experiment with different combinations of predictors, includ- 
#ing possible transformations and interactions, for each of the methods. 
#Report the variables, method, and associated confu- sion matrix that appears to 
#provide the best results on the held out data. Note that you should also experiment with values for K in the KNN classifier.

threshold=.10

train=subset(Weekly, Year>1989|Year<2009)
test=subset(Weekly, Year<1990|Year>2008)
train.lr=glm(Direction ~ Lag1 + Lag2, data=o, family="binomial")
test_ind=nrow(test)
train.lr$predict_prob=predict(train.lr, test)
train.lr$predict_dir=rep("Down", test_ind)
train.lr$predict_dir[train.lr$predict_prob>threshold]="Up"
confuse_table_2=data.frame(train.lr$predict_dir, test$Direction)
colnames(confuse_table_2)=c("predicted","actual")
table(confuse_table_2)
mean(test$Direction==train.lr$predict_dir)

print("Here I used a Logistic Regression which included Lag1 and Lag2 as predictors. Additionally, I changed the threshold until I got the lowest error from the model that I could find")

volLogTrans_train=data.frame(log(train$Volume))
colnames(volLogTrans_train)="LogVol"
volLogTrans_test=data.frame(log(test$Volume))
colnames(volLogTrans_test)="LogVol"

train_log=cbind(train$Lag2,volLogTrans_train)
test_log=cbind(test$Lag2,volLogTrans_test)
print("KNN, k=1")
knn.pred_log=knn(train_log, test_log, train$Direction ,k=1)
table(knn.pred_log, test$Direction)
print("KNN, k=3")
knn.pred_log=knn(train_log, test_log, train$Direction ,k=3)
table(knn.pred_log, test$Direction)
print("KNN, k=5")
knn.pred_log=knn(train_log, test_log, train$Direction ,k=5)
table(knn.pred_log, test$Direction)

print("Here I transformed the Volume predictor using a log transform. I included it and Lag2 in three KNN models, each w different k's. As you can see, classification breaks as k increases")
```
